```python
## Removing duplicates:
```

```python
df = pd.DataFrame({
    "name":["Alice","John","Alice","Jeff"],
    "marks":[np.nan,80,70,60],
    "city":[np.nan, "New York", "San Francisco", np.nan]
})
```

```python
df
```

```python
df.drop_duplicates(subset=['name'],keep= 'first')
```

```python
df.drop_duplicates(subset=['name'],keep= 'last')
```

```python
df.drop_duplicates(subset=['name'],keep= False)
```
